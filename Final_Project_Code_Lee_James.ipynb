# Data Ingestion

```python
#Download data from Kaggle
#Import downloaded .csv data file

data = pd.read_csv('/Users/jameslee/Desktop/Metis_DS_BC/07_Data_Engineering/Project/gun-violence-data_01-2013_03-2018.csv')
```

```python
data.head()
```

# Preprocessing: Data Cleaning/Manipulating

```python
#Clean/Manipulate data by removing unecessary column
#Reduce columns down to: date, state, city_or_country, address, n_killed, n_injured, state_house_distric, and state_senate_district

df = data.drop(columns=['incident_id',
               'incident_url',
               'source_url',
               'incident_url_fields_missing',
               'gun_stolen',
               'gun_type',
               'incident_characteristics',
               'location_description',
               'n_guns_involved',
               'notes',
               'participant_age',
               'participant_age_group',
               'participant_gender',
               'participant_name',
               'participant_relationship',
               'participant_status',
               'participant_type',
               'sources'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df.date)
df.head()
```

```python
df.dtypes
```

```python
df['year']=df.date.dt.year
df.head()
```

```python
df['month']=df.date.dt.month
df.head()
```

```python
df_final = df[['year','month','state','city_or_county','n_killed','n_injured','latitude','longitude']]
df_final.head()
```

```python
df_final.info()
```

```python
#Dropping rows with non-numeric latitude and longitude columns
df_final = df_final.dropna(subset=["latitude","longitude"])
df_final.head()
```

```python
df_final.info()
```
